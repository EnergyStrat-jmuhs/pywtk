```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas
import sys
from pywtk.wtk_api import get_wind_data, get_forecast_data
```

```python
start = pandas.Timestamp('2007-08-01', tz='utc')
end = pandas.Timestamp('2007-08-02', tz='utc')
sys.stdout.write("Loading met data [")
sys.stdout.flush()
site_data = {}
for site_id in ["1337"]:
    site_data[site_id] = get_wind_data(site_id, start, end, attributes=["power"])
    sys.stdout.write(site_id + ", ")
    sys.stdout.flush()
print "]"
```

```python
site_data["1337"].plot()
```

```python
attributes = ['hour_ahead_power', 'hour_ahead_power_p10', 'hour_ahead_power_p90']
fcst_data = {}
for site_id in ["1337"]:
    fcst_data[site_id] = get_forecast_data(site_id, start, end, attributes=attributes)
```

```python
fcst_data["1337"].plot()
```
