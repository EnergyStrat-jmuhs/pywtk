### Load interesting sites

```python
import csv
import json
site_data = []
with open('interesting_sites.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    max_lat = None
    min_lat = None
    max_lon = None
    min_lon = None
    for row in reader:
        mapped_row = map(float, (row['lat'], row['lon'], row['capacity'], row['capacity_factor']))
        if min_lat is None:
            max_lat = mapped_row[0]
            min_lat = mapped_row[0]
            max_lon = mapped_row[1]
            min_lon = mapped_row[1]
        site_data.append(mapped_row)
        max_lat = max(max_lat, mapped_row[0])
        min_lat = min(min_lat, mapped_row[0])
        max_lon = max(max_lon, mapped_row[1])
        min_lon = min(min_lon, mapped_row[1])
        #if len(site_data) > 1000:
        #    break
    #print("Bounds are %s %s %s %s"%(min_lat, max_lat, min_lon, max_lon))
    center = {"lat": (max_lat + min_lat) / 2, "lng": (max_lon + min_lon) / 2}
print("Center is %s"%center)
print("Have %s sites"%len(site_data))
# Store as javascript object in window
#Javascript("window.site_data={}".format(json.dumps(site_data)))
```
