```python
%matplotlib inline
import matplotlib.pyplot as plt
import os
import pandas
import sys
from pywtk.wtk_api import get_nc_data_from_file
```

```python
filename = os.path.join(WIND_MET_NC_DIR,"126","63311.nc")
#attributes = ["power", "wind_direction", "wind_speed", "temperature",
#              "pressure", "density"]
attributes = ["power", "wind_speed"]
leap_day = False
utc = False
start = pandas.Timestamp('2011-01-01', tz='America/Denver')
end = pandas.Timestamp('2011-01-15', tz='America/Denver')
wind_data = get_nc_data_from_file(filename, start, end, attributes=attributes, leap_day=leap_day, utc=utc)
```
